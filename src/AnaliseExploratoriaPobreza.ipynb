### Matrix de Correlação
- O coeficiente de correlação quantifica o grau e a direção da relação linear entre duas variáveis. **Ele varia de -1 a 1, onde -1 indica uma correlação negativa perfeita, 1 indica uma correlação positiva perfeita e 0 indica nenhuma correlação linear.**
- **Identificar Relacionamentos:** Ela ajuda a identificar quais variáveis têm correlações fortes e quais têm correlações fracas. Isso é útil para entender quais variáveis podem estar relacionadas e, portanto, podem afetar umas às outras.

```python
# Calcule a matriz de correlação
df_MUN_tratada = df_MUN.copy()

#Dropando Colunas Inúteis
colunas_a_descartar = ['cod_mun', 'cod_regiao', 'cod_uf', 'nom_mun']
ddf_MUN_tratada = df_MUN_tratada.drop(colunas_a_descartar, axis=1, inplace=True)

#Fazendo a Matrix de Correlação
correlation_matrix = df_MUN_tratada.corr(numeric_only=True)
```

```python
plt.figure(figsize=(30, 24))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

- Para fins de análise, escolher só um métrica das várias correlacionadas pode ser bem útil.
    - Métrica de pobreza: só uma (domicílio ou pessoa).
    - IPM só um.
    - Submétricas em funções das maiores, escolher elas ou as agregadas.
